For each of the numeric attributes (proteins upto vitamins), identify and replace all missing data(indicated with -1) with the arithmetic mean of the attribute.

```python
nattr = ['protein', 'fat', 'sodium', 'fiber', 'carbo', 'sugars', 'potass', 'vitamins']

for attr in nattr:
    df[attr] = pd.to_numeric(df[attr], errors='coerce')
    df[attr] = df[attr].replace(-1, np.nan)
    df[attr] = df[attr].fillna(df[attr].mean(skipna=True))

df
```

```python
nattr = ['protein', 'fat', 'sodium', 'fiber', 'carbo', 'sugars', 'potass', 'vitamins']

for attr in nattr:
    df[attr] = pd.to_numeric(df[attr], errors='coerce')
    df[attr] = df[attr].fillna(df[attr].median(skipna=True))

df
```

Create a CSV file called “CustomerPurchases.csv” containing details of 10 customers: Customer ID, Age, Gender, Product Purchased, Quantity, Price.

```python
products = ['Smartphone', 'Laptop', 'Tablet', 'TV', 'Refrigerator', 'Washing machine', 'Dishwasher']

data = {
    'Customer ID': list(range(1, 11)),
    'Age': np.random.randint(18, 50, size=10),
    'Gender': np.random.choice(['M', 'F'], size=10),
    'Product Purchased': np.random.choice(products, size=10),
    'Quantity': np.random.randint(1, 5, size=10),
    'Price': np.random.randint(100, 10000, size=10)
}

df = pd.DataFrame(data)
df.set_index('Customer ID', inplace=True)
df.to_csv('CustomerPurchases.csv')
df
```

Read the CSV file into a data frame and find the total amount spent by customers over the entire dataset.

```python
df['Amount'] = df['Quantity'] * df['Price']
print(f"Total amount spent by customers: {df['Amount'].sum()}")
```

Write the details of all customers who purchased a specific product (e.g., "Laptop") into a file “LaptopPurchases.csv”.

```python
df_laptop = df[df['Product Purchased'] == 'Laptop']
df_laptop
```

```python
df_laptop.to_csv('LaptopPurchases.csv')
```

Handle missing values by replacing them with the mode of the respective attribute (e.g., replace missing "Age" with the mode of "Age").

```python
for column in df.columns:
    mode = df[column].mode()
    if not mode.empty:
        mode_to_fill = mode[0]
        df[column] = df[column].replace(to_replace=np.nan, value=mode_to_fill)
    else:
        if df[column].dtype == 'object':
            df[column] = df[column].replace(to_replace=np.nan, value='NA')
        elif pd.api.types.is_numeric_dtype(df[column]):
            df[column] = df[column].replace(to_replace=np.nan, value=0)
        else:
            df[column] = df[column].replace(to_replace=np.nan, method='ffill')

df
```

```python
df.describe()[3:]
```

Identify any outliers in the salary data using the interquartile range (IQR) method, and replace them with the nearest valid value.

```python
q1 = df['Salary'].quantile(0.25)
q3 = df['Salary'].quantile(0.75)
iqr = q3 - q1

upper_bound = q3 + 1.5 * iqr
lower_bound = q1 - 1.5 * iqr
outliers = df[(df['Salary'] < lower_bound) | (df['Salary'] > upper_bound)]

print("Outliers in Salary before replacement:")
if not outliers.empty:
    print(f"{outliers}\n")
else:
    print("No outliers found.\n")

for index, row in outliers.iterrows():
    if row['Salary'] > upper_bound:
        df.loc[index, 'Salary'] = upper_bound
    elif row['Salary'] < lower_bound:
        df.loc[index, 'Salary'] = lower_bound

print("Data after replacement:")
df
```

After handling the outliers, create a table with the 5-number summary for the "Salary" attribute and analyze whether the strategy was effective.

```python
df['Salary'].describe()[3:]
```
