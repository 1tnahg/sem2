How does the sugar content vary across different brands?

What is the average calories in cereals per manufacturer?

```python
plt.figure(figsize=(8, 6))
plt.bar(x=df.groupby('mfr')['calories'].mean().index, height=df.groupby('mfr')['calories'].mean().values)
plt.title('Average calories in cereals by manufacturer')
plt.xlabel('Manufacturer (mfr)')
plt.ylabel('Average calories')
plt.xticks(df.groupby('mfr')['calories'].mean().index)
plt.show()
```

What is the average nutritional content (calories, sugars, protein, fiber, etc.) across all cereals?

```python
df[['calories', 'sugars', 'protein', 'fiber', 'fat', 'sodium', 'potass']].mean()
```

What is the relationship between sugar and calories?

```python
plt.figure(figsize=(8, 6))
plt.scatter(x=df['sugars'], y=df['calories'])
plt.title('Calories vs sugar content')
plt.xlabel('Sugar content')
plt.ylabel('Calories')
plt.show()
```

How does the sugar content compare in cereals with high vs. low ratings?

```python
def categorize_rating(rating):
    if rating >= df['rating'].median():
        return 'High Rating'
    else:
        return 'Low Rating'

df['rating_category'] = df['rating'].apply(categorize_rating)

plt.figure(figsize=(6, 5))
plt.boxplot([df[df['rating_category'] == 'High Rating']['sugars'],
             df[df['rating_category'] == 'Low Rating']['sugars']],
            labels=['High Rating', 'Low Rating'])
plt.title('Sugar content distribution for high vs. low rated cereals')
plt.xlabel('Rating category')
plt.ylabel('Sugar content')
plt.tight_layout()
plt.show()
```

Do cereals marketed as "healthy" (e.g., high fiber, low sugar) tend to have higher ratings?

```python
def categorize_health(row):
    if row['fiber'] >= df['fiber'].median() and row['sugars'] <= df['sugars'].median():
        return 'Likely Healthy'
    else:
        return 'Likely Less Healthy'

df['health_category'] = df.apply(categorize_health, axis=1)

plt.figure(figsize=(6, 5))
plt.boxplot([df[df['health_category'] == 'Likely Healthy']['rating'],
             df[df['health_category'] == 'Likely Less Healthy']['rating']],
            labels=['Likely Healthy', 'Likely Less Healthy'])
plt.title('Cereal ratings for likely healthy vs. likely less healthy cereals')
plt.xlabel('Health Category')
plt.ylabel('Rating')
plt.show()
```

Is there a relationship between rating and display shelf?

```python
plt.figure(figsize=(8, 6))
plt.boxplot([df[df['shelf'] == shelf]['rating'] for shelf in df['shelf'].unique()],
            labels=df['shelf'].unique())
plt.title('Distribution of cereal ratings by display shelf')
plt.xlabel('Display shelf')
plt.ylabel('Rating')
plt.show()
```
