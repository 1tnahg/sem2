```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('mtcars.csv')
df.head()
```

Extract the 'mpg' column from the dataframe & Plot the mpg values as a line plot.

```python
df['mpg'].plot()
```

Select only the 'cyl', 'wt', and 'mpg' columns from the dataframe & Plot the selected dataframe columns.

```python
df[['cyl', 'wt', 'mpg']].plot()
```

Bar Plot - the count of cars by the number of cylinders.

```python
df['cyl'].value_counts().plot(kind='bar', color='green')

plt.title('Count of cars by cylinders')
plt.xlabel('Cylinders')
plt.ylabel('Count')
plt.show()
```

Scatter plot of weight vs mpg.

```python
plt.figure(figsize=(8, 6))
plt.scatter(df['wt'], df['mpg'], color='red')

plt.title('Scatter plot of fuel efficiency vs weight')
plt.xlabel('Weight')
plt.ylabel('Fuel efficiency (mpg)')
plt.show()
```

Box plot of mpg for each cylinder category.

```python
plt.figure(figsize=(8, 6))
plt.boxplot([df[df['cyl'] == 4]['mpg'], df[df['cyl'] == 6]['mpg'], df[df['cyl'] == 8]['mpg']])

plt.title('Box plot of fuel efficiency by number of cylinders')
plt.xlabel('Number of cylinders')
plt.ylabel('Fuel efficiency (mpg)')
plt.xticks([1, 2, 3], ['4', '6', '8'])
plt.show()
```

Histogram of mpg for all cars.

```python
plt.hist(df['mpg'], bins=5, color='hotpink', edgecolor='black')

plt.title('Histogram of fuel efficiency')
plt.xlabel('Fuel efficiency (mpg)')
plt.ylabel('Frequency')
plt.show()
```

This pie chart shows the proportion of cars with 4, 6, or 8 cylinders.

```python
plt.figure(figsize=(8, 6))
df['cyl'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=['lightblue', 'lightcoral', 'lavender'])

plt.title('Proportion of Cars by Number of Cylinders')
plt.ylabel('Number of cylinders')
plt.show()
```

Pie Chart for Transmission Types.

```python
df['am'] = df['am'].map({0: 'Automatic', 1: 'Manual'})
plt.figure(figsize=(8, 6))
df['am'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=['lightblue', 'lightcoral'])

plt.title('Proportion of Cars by Transmission Type')
plt.ylabel('')
plt.show()
```

Nested Pie Chart which needs Data preparation.

```python
outer_labels = df['cyl'].value_counts().index
outer_sizes = df['cyl'].value_counts().values

inner_labels = df['gear'].value_counts().index
inner_sizes = df['gear'].value_counts().values

# Create nested pie chart
fig, ax = plt.subplots(figsize=(8, 8))
ax.pie(outer_sizes, labels=outer_labels, radius=1, autopct='%1.1f%%', colors=['lightblue', 'lightcoral', 'lavender'])
ax.pie(inner_sizes, labels=inner_labels, radius=0.7, autopct='%1.1f%%', colors=['hotpink', 'darkblue', 'red'])
ax.set(aspect="equal")
plt.title('Nested Pie Chart: Cylinders (Outer) and Gears (Inner)')
plt.show()
```
